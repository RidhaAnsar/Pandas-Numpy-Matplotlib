```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.plot(x="age", y="trestbps", kind="scatter");
```

```python
heart_disease.plot(x="age", y="trestbps", kind="hist", bins=20)
```

```python
heart_disease.plot.hist(subplots=True);
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,30));
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
len(over_50)
```

```python
over_50.plot(x="age", y="chol", c="target", kind="scatter");
```

```python
fig, ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
ax.set(title="Heart disease and cholestrol level",
       xlabel="AGE",
       ylabel="CHOLESTROL");
```
